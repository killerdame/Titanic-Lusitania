```python
#Cleaning up the Age column of the Lusitania Data
lusitania_db['Age'] = lusitania_db['Age'].str.replace(' \?','.0')
lusitania_db['Age'] = lusitania_db['Age'].str.replace('(\d\d)\?','\\1')
lusitania_db['Age'] = lusitania_db['Age'].str.replace('(\d)\?','\\1')
lusitania_db['Age'] = lusitania_db['Age'].str.replace('(...\d\d)','\\1')
lusitania_db['Age'] = lusitania_db['Age'].str.replace('\?','30')
lusitania_db['Age'] = lusitania_db['Age'].str.replace('2_','20')
lusitania_db['Age'] = lusitania_db['Age'].str.replace('..-months','1')
lusitania_db['Age'] = lusitania_db['Age'].str.replace('Infant','1')
lusitania_db['Age'] = lusitania_db['Age'].str.replace('45 or 51','47')
lusitania_db['Age'] = lusitania_db['Age'].str.replace('51 or 47','49')
lusitania_db['Age'] = lusitania_db['Age'].str.replace('(\d\d)..(\d\d).','\\1')
lusitania_db['Age'] = lusitania_db['Age'].str.replace('(\d)..(\d).','\\1')
# lusitania_db
lusitania_db['Age'].unique() 
```

```python
#Converting the Age column of the Lusitania db to a float
lusitania_db['Age'] = lusitania_db['Age'].astype(float)
```

```python
#Dropping lusitania columns not needed for analysis
lusitania_db.drop(["Citizenship"], axis=1, inplace=True)
lusitania_db.drop(["Position"], axis=1, inplace=True)
lusitania_db.drop(["Lifeboat"], axis=1, inplace=True)
lusitania_db.drop(["Rescue Vessel"], axis=1, inplace=True)
lusitania_db.drop(["Value"], axis=1, inplace=True)
lusitania_db.drop(["Family name"], axis=1, inplace=True)
lusitania_db.drop(["Title"], axis=1, inplace=True)
lusitania_db.drop(["Personal name"], axis=1, inplace=True)
lusitania_db.drop(["Adult/Minor"], axis=1, inplace=True)
lusitania_db.drop(["Body No."], axis=1, inplace=True)
lusitania_db.drop(["County"], axis=1, inplace=True)
lusitania_db.drop(["State"], axis=1, inplace=True)
lusitania_db.drop(["City"], axis=1, inplace=True)
lusitania_db.drop(["Status"], axis=1, inplace=True)
lusitania_db.drop(["Country"], axis=1, inplace=True)
lusitania_db.drop(["Ticket No."], axis=1, inplace=True)
lusitania_db.drop(["Cabin No."], axis=1, inplace=True)
lusitania_db.drop(["Traveling Companions and other notes"], axis=1, inplace=True)
# Combining the Saved and Saved (Died from trauma) column into one column and 
# the Lost and Not on board columns into one column
lusitania_db["Fate"] = lusitania_db["Fate"].str.replace("Saved \(died from trauma\)", "Saved")
lusitania_db["Fate"].replace("Not on board", "Lost", inplace=True)

# Consolidating and renaming the Class column to be able to compare to Titanic DB
lusitania_db["Department/Class"].replace('Saloon', '1', inplace=True)
lusitania_db["Department/Class"].replace('Second', '2', inplace=True)
lusitania_db["Department/Class"].replace('Third', '3', inplace=True)
lusitania_db["Department/Class"].replace('Third (Distressed British Seaman)', '3', inplace=True)
```

```python
#Changing the "Fate" column to a Boolean and changing the names 
lusitania_db["Saved"] = lusitania_db["Fate"].apply(lambda x: 1 if x == "Saved" else 0)
lusitania_db["Lost"] = lusitania_db["Fate"].apply(lambda x: 1 if x == "Lost" else 0) 
```

```python
# Drop titanic columns not needed for analysis
titanic_db.drop(["boat"], axis=1, inplace=True)
titanic_db.drop(["body"], axis=1, inplace=True)
titanic_db.drop(["name"], axis=1, inplace=True)
titanic_db.drop(["sibsp"], axis=1, inplace=True)
titanic_db.drop(["parch"], axis=1, inplace=True)
titanic_db.drop(["ticket"], axis=1, inplace=True)
titanic_db.drop(["fare"], axis=1, inplace=True)
titanic_db.drop(["cabin"], axis=1, inplace=True)
#Create a boolean for the Survived column
titanic_db["deceased"] = titanic_db["survived"].apply(lambda x: 1 if x == 0 else 0)
```

#### For the Age of the passangers ... how would you guess at the missing values using the other data present in the CSV.

```python
titanic_db["age"].fillna(titanic_db.groupby("sex")["age"].transform("mean"), inplace=True)

titanic_db['age'].unique()
lusitania_db['Age'].unique()
```

```python
#Taking the missing data from the Age column and grouping it by sex and class and then replacing it with the 
#mean of each group. 

titanic_db['age'] = titanic_db.groupby(['sex', 'pclass'])['age'].transform(lambda x: x.fillna(x.mean()))
titanic_db
```

```python
#Taking the missing data from the Age column and grouping it by sex and class and then replacing it with 
#the mean of each group. 
lusitania_db['Age'] = lusitania_db.groupby(['Sex', 'Department/Class'])['Age'].transform(lambda x: x.fillna(x.mean()))
lusitania_db['Age'].unique()     
```
